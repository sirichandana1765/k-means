# 📊 Customer Analytics: Exploratory Data Analysis & Elbow Analysis

This notebook demonstrates exploratory data visualization on the Mall Customers dataset and identifies the optimal number of clusters for consumer segmentation using the **Elbow Method (WCSS)**.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Set seaborn style
sns.set_theme(style="darkgrid")
```

## 1. Load data and Inspect Schema

```python
df = pd.read_csv('../data/datasets.csv')
df.head()
```

```python
df.info()
```

## 2. Demographic Exploratory Plots

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.histplot(data=df, x='Age', kde=True, color='skyblue')
plt.title('Distribution of Age')

plt.subplot(1, 3, 2)
sns.histplot(data=df, x='Annual Income (k$)', kde=True, color='teal')
plt.title('Distribution of Income')

plt.subplot(1, 3, 3)
sns.histplot(data=df, x='Spending Score (1-100)', kde=True, color='magenta')
plt.title('Distribution of Spending Score')

plt.tight_layout()
plt.show()
```

## 3. Bivariate Relationships (Income vs Spending Score)

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender', s=80, alpha=0.8)
plt.title('Annual Income vs Spending Score Scatterplot')
plt.show()
```

## 4. The Elbow Curve Method to find Optimal K

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, random_state=42, n_init='auto')
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8, 4.5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--', color='indigo', linewidth=2)
plt.title('The K-Means Elbow Method (WCSS)', fontsize=14)
plt.xlabel('Number of Clusters (k)', fontsize=12)
plt.ylabel('Inertia (Within-Cluster Sum of Squares)', fontsize=12)
plt.xticks(range(1, 11))
plt.show()
```

### Observation:
The elbow transition is clearly prominent at **k=5**. This identifies the optimal baseline for user segmentation, dividing our users into 5 distinct classes.
